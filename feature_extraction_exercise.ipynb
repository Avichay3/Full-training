```First we would like to extract features regarding a single question. In order to do so, first create a dataset containing  all the questions (and their id. why should we remember the id?), without duplicates. Name it 'questions'.```

````python
q1 = data[['qid1', 'question1']].rename(
    columns={'qid1': 'question_id', 'question1': 'question'}
)

q2 = data[['qid2', 'question2']].rename(
    columns={'qid2': 'question_id', 'question2': 'question'}
)

questions = pd.concat([q1, q2], axis=0)

questions = questions.drop_duplicates(subset='question_id')

questions = questions.reset_index(drop=True)

questions.head()
````
